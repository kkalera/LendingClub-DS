## Statistical Analysis

For the statistical analysis section of this project, I'd like to explore the `DAYS_EMPLOYED` feature.

&nbsp;

**Null Hypothesis:** 

There is no statistically significant difference between the duration of employment between clients with and without repayment issues.

**Alternative Hypothesis:** 

There is a statistically significant difference between the duration of employment between clients with and without repayment issues.



&nbsp;

### Population Creation

Let's start off by creating the two populations.

```python
df_eda.loc[df_eda["DAYS_EMPLOYED"] > 350000, "DAYS_EMPLOYED"] = 0
pop_a = pd.DataFrame(df_eda.loc[df_eda["TARGET"] == 0, "DAYS_EMPLOYED"])
pop_b = pd.DataFrame(df_eda.loc[df_eda["TARGET"] == 1, "DAYS_EMPLOYED"])

logger.info(f"Population A: {pop_a.shape[0]}")
logger.info(f"Population B: {pop_b.shape[0]}")
```

&nbsp;

### Distribution Check

Let's check if the data is normally distributed. Since the population size is very large, most statistical tests for normality will tend to reject that the distribution is normally distributed. Because of this, I'll use a visual analysis tool, in this case a QQ-plot.

```python
fig = pplot(
    df_eda,
    x="DAYS_EMPLOYED",
    y=stats.gamma,
    hue="TARGET",
    kind="qq",
    height=4,
    aspect=2,
    display_kws={"identity": True},
)
plt.gcf().axes[0].set_title("QQ Plot of Days Employed")
plt.gcf().axes[0].set_xlabel("Normally Distributed Quantiles")
```

We can see that the data is not normally distributed. Let's choose an appropriate test.

&nbsp;

### Test Choice

Because of the large sample sizes and non-uniform distribution of the populations, I'm opting for the Mann-Whitney U-Test.

&nbsp;

**Significance Level:** 0.05


```python
u_stat, p_value = stats.mannwhitneyu(pop_a, pop_b, alternative="two-sided")
logger.info(f"Mann-Whitney U statistic: {u_stat}, p-value: {p_value}")
if p_value < 0.05:
    logger.info("Reject the null hypothesis")
else:
    logger.info("Failed to reject the null hypothesis")
```

### Confidence Intervals

```python
reload()
ci_a = data_f.get_confidence_interval(pop_a["DAYS_EMPLOYED"])
ci_b = data_f.get_confidence_interval(pop_b["DAYS_EMPLOYED"])

logger.info(
    f"Population A: {ci_a[0]:.2f} , {ci_a[1]:.2f}  --  Delta: {ci_a[1] - ci_a[0]:.2f}"
)
logger.info(
    f"Population B: {ci_b[0]:.2f} , {ci_b[1]:.2f}  --  Delta: {ci_b[1] - ci_b[0]:.2f}"
)
```

### Summary

The analysis shows that there is a statistically significant difference between the employment length of applicant that experienced repayment issues and those that did not.

The significance level was set to 0.05 and confidence intervals showed that the spread was larger for population B (those that experienced repayment issues), but was still relatively small.

### Product 1: Anomaly Detection In Applications

For anomaly detection, I'm going to use an unsupervised isolation forest model. The goal of the anomaly detection is to identify anomaly's as a sort of alarm system that these applications need to be inspected by a human instead of relying on the automated system. To be clear: this system will look at whether or not the loan experienced repayment issues.

&nbsp;

Before diving straight into developing the model, there's some important considerations to address.

- **Data Dimensionality:** High dimensionality can be challenging for anomaly detection. To solve this, principal component analysis will be used to reduce the amount of features used in this step.

- **Null Values:** Isolation forests do not handle null values well. We will need to handle them before doing PCA and modeling.

```python
reload()
results = ml_f.run_pca(df_eda.drop(columns=["TARGET"]), n_components=10)
results
```

These are the columns that arose from the PCA analysis. These will be used for training the isolation forest.

```python
df_eda[results].head()
```

&nbsp;

#### Model Training

You might be thinking about the fact that I stated earlier that the null values need to be handled. This is handled in the pipeline, here's an overview of what is done in the pipeline:

- Null values are imputed using a SimpleImputer.
- Outliers are clipped to the interquartile range.
- Values are scaled to a standard deviation of 1 and mean of 0.
- Values are normalized to a range of 1.

```python
builder = ml_f.PipeBuilder()
anomaly_model_path = "src/models/anomaly_pipe.pkl"


if not os.path.exists(anomaly_model_path):

    pipe = builder.build(
        model=IsolationForest(random_state=42, n_jobs=-1, contamination=0.0796),
        num_columns=results,
        num_steps=[
            builder.steps["simple_imputer_median"],
            builder.steps["iqr_scaler"],
            builder.steps["standard_scaler"],
            builder.steps["normalizer"],
        ],
    )
    pipe.fit(df_eda[results])
    joblib.dump(pipe, anomaly_model_path)

else:

    pipe = joblib.load(anomaly_model_path)


df_eda["anomaly"] = pipe.predict(df_eda[results])
df_eda["anomaly"] = df_eda["anomaly"].astype("category")
df_eda["anomaly"] = df_eda["anomaly"].cat.rename_categories({1: "Normal", -1: "Anomaly"})
vc = df_eda.groupby("TARGET")["anomaly"].value_counts(normalize=True)
ratio = vc[1] / vc[0]

fig = plot_f.plot_distribution_and_ratio(
    df_eda, column_name="anomaly", ratio=ratio, hue="TARGET"
)
```

The ratio of loans that experienced payment issues is larger in the anomaly group compared to the normal group.

There's two things that this model could be useful for. First of all, the results from this model can be used as an input for the credit risk assessment model that will be created later. 

Second, the results from this could trigger a system that makes sure that this application is reviewed by a human before accepting or declining the loan. If we were to rely on this system to decide which applications to automate, that could potentially automate a large portion of loans.

Before moving on to the next product, let's take a look at the differences between applications flagged as anomalous and normal.

&nbsp;

#### Model Examination

We already looked at ratio of loans that experienced issues among anomalies, let's take a look at the other columns that were used to train the model. For visualization purposes, these values will be normalized.

```python
df_norm = df_eda[results].copy()
df_norm[results] = pipe.named_steps["preprocessor"].transform(df_norm)
df_norm = pd.concat([df_norm, df_eda["anomaly"]], axis=1)

df_long = (
    df_norm.groupby("anomaly", observed=False)
    .mean()
    .reset_index()
    .melt(id_vars="anomaly", var_name="Feature", value_name="Mean")
)

plt.figure(figsize=(10, 6))
sns.barplot(
    data=df_long, y="Feature", x="Mean", hue="anomaly", palette=plot_f.PALETTE[-1::-1]
)
plt.title("Mean Values of Features for Anomaly vs Normal")
plt.show()
```

The biggest differences between normal and anomalous loans is the open credit, total credit and the closed credit. These are the features we engineered and seem to be a good indicator to separate anomalous applications from the mean.
